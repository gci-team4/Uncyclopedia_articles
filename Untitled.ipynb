```python
import re

file_name = "ja-pages_current.xml"
with open(file_name) as datafile:
    allresult = []
    i = 0
    for line in datafile:
        i += 1
        pattern = re.compile("<title>(.*)</title>")
        result = pattern.findall(line)
        if len(result) > 0:
            allresult.append(result[0])
        if i % 100000 == 0:
            print(i, "rows done")
    print(i, "rows finished")
```
